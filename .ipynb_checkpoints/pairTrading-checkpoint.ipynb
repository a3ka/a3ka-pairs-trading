## II) Нормальное распределение

### 1) Введение

Самым простым показателем работоспособности спреда пары активов будет наличие в движении цен нормального распределения, которое, для напоминание, выгляет как на изображение ниже.  

```python
# Параметры нормального распределения
mu = 0  # среднее значение
sigma = 1  # стандартное отклонение

# Создаем массив значений для оси X
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
# Вычисляем плотность нормального распределения
y = norm.pdf(x, mu, sigma)

# Создаем график
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='blue', lw=2)

# Область в пределах двух стандартных отклонений
plt.fill_between(x, y, where=((x >= mu - 2*sigma) & (x <= mu + 2*sigma)), 
                 color='lightblue', alpha=0.5, label='95% within 2 standard deviations')

# Линии среднего и стандартных отклонений
plt.axvline(mu, color='red', linestyle='--', lw=2, label='Mean (μ)')
plt.axvline(mu - 2*sigma, color='black', linestyle='-', lw=1.5)
plt.axvline(mu + 2*sigma, color='black', linestyle='-', lw=1.5)

# Настройки графика
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Отображаем график
plt.show()
```

Проверка спреда на нормальное распределение в парном трейдинге важна по нескольким причинам. Это помогает понять, насколько хорошо спред подходит для стратегий реверсии к среднему и для оценки вероятностей возможных движений спреда. Вот подробное объяснение, почему это важно:

#### 1. **Предсказуемость и реверсия к среднему**:
   - **Стратегии реверсии к среднему** предполагают, что спред между активами имеет тенденцию возвращаться к своему среднему значению после отклонений.
   - Если распределение спреда близко к нормальному, это означает, что его отклонения от среднего распределяются предсказуемо. В случае нормального распределения большинство значений (около 68%) находятся в пределах одного стандартного отклонения от среднего.
   - Знание того, что спред нормально распределен, помогает лучше понимать вероятности возврата к среднему. Например, если спред отклонился на 2 стандартных отклонения от среднего, вероятность того, что он вернется, выше, если распределение нормальное.

#### 2. **Оценка риска и волатильности**:
   - **Нормальное распределение** — это также стандартный метод в финансовой математике для оценки рисков, потому что оно позволяет использовать известные методы для оценки вероятностей.
   - Например, в нормальном распределении мы знаем, что вероятность отклонения на 3 стандартных отклонения от среднего очень мала (около 0.3%). Это помогает трейдерам оценить риски при создании торговых стратегий.
   - Если спред не имеет нормального распределения, это может указывать на наличие более частых резких движений (высокая хвостовость распределения) или асимметрию, что требует осторожности при оценке рисков.

#### 3. **Моделирование уровней открытия и закрытия позиций**:
   - Когда спред имеет нормальное распределение, трейдеры (то есть мы) могут использовать параметры распределения (среднее и стандартное отклонение) для определения уровней входа и выхода из позиций.
   - Например, можно настроить стратегию таким образом, чтобы открывать позиции, когда спред отклоняется на 2 стандартных отклонения от среднего, и закрывать их, когда он возвращается к среднему.
   - Если спред отклоняется от нормальности (например, имеет длинные хвосты или сильно асимметричен), это может сделать такие стратегии менее эффективными, так как вероятность сильных отклонений от среднего выше, чем предполагается в нормальном распределении.

#### 4. **Проверка корректности модели и гипотез**:
   - В парном трейдинге часто предполагается, что активы связаны линейной зависимостью, а спред между ними является стационарным. Но стационарность не обязательно означает нормальность.
   - Проверка на нормальное распределение помогает понять, насколько эта гипотеза корректна. Если спред сильно отклоняется от нормальности, это может быть сигналом к пересмотру модели или поиску других подходящих активов для торговли.
   - Например, если спред имеет асимметричное распределение или длинные хвосты, это может указывать на структурные особенности рынка или влияние внешних факторов, которые могут нарушить предположения вашей модели.

#### 5. **Нормальность и методы статистического анализа**:
   - Многие статистические методы и тесты (включая тест Дики-Фуллера) предполагают нормальность временных рядов. Если спред близок к нормальному распределению, результаты таких тестов будут более надежными.
   - Это особенно важно, если вы используете статистические методы для прогнозирования будущих движений или оценки вероятностей.

#### Как проверить нормальность:
Для проверки нормальности можно использовать такие тесты, как **тест Шапиро-Уилка**, **тест Жарка-Бера**, или просто построить **гистограмму** и **Q-Q график** (квантиль-квантиль график), чтобы визуально оценить, насколько распределение спреда близко к нормальному.

Перейдем тогда к анализу спредов на "нормальность" рассматриваемых нами пар, и начем с визуализации их спредов и графиков для визуального анализа.

### 3) Математическое тестирование на нормальность распределения

**Тест Шапиро-Уилка**

Тест Шапиро-Уилка доступен в библиотеке *scipy*. Нулевая гипотеза предполагает, что распределение данных является нормальным. Если p-значение больше выбранного уровня значимости (p-уровня), мы предполагаем, что распределение нормальное. В противном случае делаем вывод, что распределение не является нормальным.

```python
def is_normal(sample, test=stats.shapiro, p_level=0.05, **kwargs):
    """Apply a normality test to check if sample is normally distributed."""
    t_stat, p_value = test(sample, **kwargs)
    print("Test statistic: {}, p-value: {}".format(t_stat, p_value))
    print("Is the distribution Likely Normal? {}".format(p_value > p_level))
    return p_value > p_level
```

```python
# Using Shapiro-Wilk test (default)
print("STRK/ZK:-"); is_normal(price_ratio_strk_zk);
print("STRK/ENA:-"); is_normal(price_ratio_strk_ena);
print("STRK/W:-"); is_normal(price_ratio_strk_w);
```

**Тест Колмогорова-Смирнова**

Тест Колмогорова-Смирнова доступен в библиотеке *scipy.stats*. Этот тест сравнивает распределение данных с теоретическим распределением. Мы выберем нормальное распределение ('norm') в качестве теоретического и зададим среднее и стандартное отклонение для этого распределения. Среднее и стандартное отклонение теоретической нормы будут равны среднему и стандартному отклонению фактического распределения данных.

```python
def is_normal_ks(sample, test=stats.kstest, p_level=0.05, **kwargs):
    """
    sample: a sample distribution
    test: a function that tests for normality
    p_level: if the test returns a p-value > than p_level, assume normality
    
    return: True if distribution is normal, False otherwise
    """
    normal_args = (np.mean(sample), np.std(sample))
    
    t_stat, p_value = test(sample, 'norm', normal_args, **kwargs)
    print("Test statistic: {}, p-value: {}".format(t_stat, p_value))
    print("Is the distribution Likely Normal? {}".format(p_value > p_level))
    return p_value > p_level
```

```python
# Using Shapiro-Wilk test (default)
print("STRK/ZK:-"); is_normal_ks(price_ratio_strk_zk);
print("STRK/ENA:-"); is_normal_ks(price_ratio_strk_ena);
print("STRK/W:-"); is_normal_ks(price_ratio_strk_w);
```

## II) Стационарность спреда

**Стационарность** — это свойство временного ряда, при котором его статистические характеристики (например, среднее, дисперсия) не изменяются со временем. Основные признаки стационарного ряда:
- Постоянное среднее.
- Постоянная дисперсия (волатильность).
- Отсутствие тренда (временная зависимость между значениями ряда ограничена).
Стационарные временные ряды проще анализировать и прогнозировать, так как их поведение не меняется в зависимости от времени.

**Стационарность спреда** означает, что отношение между двумя временными рядами, например, ценами двух активов, остается стабильной во времени и не демонстрирует тенденции к систематическому росту или падению.

Если спред стационарен:
- Он колеблется вокруг своего среднего значения.
- Имеет определенную степень предсказуемости (возвращается к среднему значению).
- Это может свидетельствовать о наличии долгосрочной равновесной связи между двумя активами. Например, цены двух акций могут быть связаны между собой через экономические или рыночные факторы, и их спред со временем стремится вернуться к некоторому равновесному уровню.

Если спред не стационарен:
- Он будет демонстрировать тренд, то есть может увеличиваться или уменьшаться с течением времени.
- Это может указывать на то, что между активами нет устойчивой долгосрочной связи, и разница между их ценами может увеличиваться без возврата к среднему уровню.
  
**Зачем проверять стационарность спреда?**
Проверка стационарности спреда важна в парном трейдинге и других стратегиях, где предполагается, что спред между двумя активами будет возвращаться к среднему значению. Если спред стационарен, то можно использовать стратегии на основе торговли отклонений спреда от среднего значения. 

Нормально распределённый и стационарный спред проще всего использовать в парном трейдинге, поскольку его цена колеблется вокруг среднего значения с предсказуемыми отклонениями.

В нашем исследовании для определения стационарности спреда мы будем использовать _*расширенный тест Дики-Фуллера (Augmented Dickey-Fuller Test)*_. Этот тест проверяет наличие единичного корня в временном ряде, что помогает определить, является ли ряд стационарным. Если тест отвергает нулевую гипотезу о наличии единичного корня, это свидетельствует о стационарности ряда, то есть его статистические свойства остаются стабильными во времени.

#### Augmented Dickey-Fuller Test

```python
def is_spread_stationary(spread, p_level=0.05):
    """
    spread: obtained from linear combination of two series with a hedge ratio
    
    p_level: level of significance required to reject null hypothesis of non-stationarity
    
    returns:
        True if spread can be considered stationary
        False otherwise
    """
    
    # Убираем NaN и бесконечные значения из спреда
    spread = spread.replace([np.inf, -np.inf], np.nan).dropna()
    
    # Используем adfuller для проверки стационарности
    adf_result = adfuller(spread, maxlag=None, regression='c', autolag='AIC', store=False, regresults=False)
    
    # Получаем p-value
    pvalue = adf_result[1]
    
    print(f"pvalue {pvalue:.4f}")
    if pvalue <= p_level:
        print(f"pvalue is <= {p_level}, assume spread is stationary")
        return True
    else:
        print(f"pvalue is > {p_level}, assume spread is not stationary")
        return False
```

```python
# Предположим, у вас уже есть рассчитанный спред:
is_stationary = is_spread_stationary(price_ratio_strk_ena)

# Результат функции (True/False) укажет, является ли спред стационарным:
if is_stationary:
    print("The spread is stationary, it may be suitable for mean-reverting strategies.")
else:
    print("The spread is not stationary, it may not be suitable for mean-reverting strategies.")
```
